```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from scipy import stats
```

#Algoritmo


```python
class KNN():
     
    def __init__(self, K):
        self.K = K
    
    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X, y):
        wrongs = rights = 0
        y_pred = []
        i = 0
        for x in X:
            predict = self._predict(x)
            y_pred.append(predict)
            if predict == y[i]:
                rights += 1
            else:
                wrongs += 1
            i += 1
        return np.array(y_pred), rights/(X.shape[0])

    def _predict(self, x):
        distances = [self.manhattan_distance(x, x_train) for x_train in self.X_train]
        k_index = np.argsort(distances)[:self.K]
        k_neighbor_labels = [self.y_train[i] for i in k_index]
        result = stats.mode(k_neighbor_labels)
        return result[0][0]

    def manhattan_distance(self, sample, centroid):
        return abs((sample[0] - centroid[0])) + abs((sample[1] - centroid[1]))
```

#Dados

```python
X, y = load_iris(return_X_y=True)
df = pd.DataFrame(data= np.c_[X, y], columns= ['sepal_lenght', 'sepal_width', 'petal_length', 'petal_width', 'target'])
df
```

##Normalizando

```python
normalized_data = StandardScaler().fit_transform(df)
normalized_data = pd.DataFrame(data=normalized_data)
normalized_data
```

##Aplicando *Principal Component Analysis* para diminuir as dimensÃµes do dataset

```python
pca = PCA(n_components=2)
x_transformed = pca.fit_transform(normalized_data)
```

#Dividindo o dataset em treino e teste

```python
x_train, x_test, y_train, y_test = train_test_split(x_transformed, y, test_size=0.2)
```

#Treino e teste



```python
model = KNN(5)
model.fit(x_train, y_train)
classifications, accuracy = model.predict(x_test, y_test)

print(classifications)
print(f'AcurÃ¡cia = {accuracy}')
```
